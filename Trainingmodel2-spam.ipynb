```python
#dependencies
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
```

```python
#read csv file
spambase_df=pd.read_csv("../content/sample_data/spamcsv.csv.csv")
spambase_df.head()
```

```python
word_counts=spambase_df["class"]
word_counts
```

```python
spambase_df.drop_duplicates(keep='last')
```

```python
spambase_df.describe
```

```python
# Split our preprocessed data into our features and target arrays
X=spambase_df.drop("class",axis=1).values
y=spambase_df["class"].values
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
     
```

```python
#create a standardscaler instances
scaler=StandardScaler()
#fit the standard scaler
X_scaler=scaler.fit(X_train)
# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
     
```

Compile, Train and Evaluate the Model


```python
# Get the input feature/ shape
X_train_scaled.shape
     
```

Neural network optimization

```python
nn_model1 = tf.keras.models.Sequential()
input_features = X_train_scaled.shape[1]

# First hidden layer
nn_model1.add(tf.keras.layers.Dense(units=20, activation='relu', input_dim=input_features))

# Second hidden layer
nn_model1.add(tf.keras.layers.Dense(units=20, activation='relu'))

# Output layer

nn_model1.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
# Check the structure of the model
nn_model1.summary()
```

```python
# Compile the model
nn_model1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model

fit_model_1 = nn_model1.fit(X_train_scaled, y_train, epochs=50)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model1.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn_model1.save('../h5_files/spamcsv_model_2.h5')
```
